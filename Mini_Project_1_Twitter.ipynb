```python
#$ sudo pip install pattern 
#$ sudo pip install matplotlib

from pattern.web import Twitter
from pattern.en import sentiment
from matplotlib import numpy as np
import matplotlib.pyplot as plt

def sent_analysis(company, n):
    count_total = 0.0
    count_positive = 0
    count_neutral = 0
    count_negative = 0
    for tweet in Twitter().search(company, count = n):
        count_total = count_total + 1
        sent = sentiment(tweet.text)
        print tweet.text, sent,
        if sent[0] > 0.1:
            print 'positive'
            count_positive = count_positive + 1
        elif -0.1 < sent[0] < 0.1:
            print 'neutral'
            count_neutral = count_neutral + 1
        else:
            print 'negative'
            count_negative = count_negative + 1
        percent_positive = (count_positive/count_total)*100
        percent_neutral = (count_neutral/count_total)*100
        percent_negative = (count_negative/count_total)*100
    print 'the number of tweets that have a positive sentiment is', count_positive, 'or', percent_positive, '%'
    print 'the number of tweets that have a neutral sentiment is', count_neutral, 'or', percent_neutral, '%'
    print 'the number of tweets that have a negative sentiment is', count_negative, 'or', percent_negative, '%'
    #code for bar graph below using matplotlib python package
    x_axis = ('Positive', 'Neutral', 'Negative')
    y_axis = np.arange(len(x_axis))
    count_sent = [count_positive, count_neutral, count_negative]
    plt.title('Sentiment Analysis Type BreakDown')
    plt.xticks(y_axis, x_axis)
    plt.ylabel('Count of Tweets')
    plt.bar(y_axis, count_sent, align = 'center')
    plt.show()

sent_analysis('@JetBlue', 100)
```
